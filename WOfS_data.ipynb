```python
from collections import namedtuple 
import numpy as np

'''bit 7 - water present Decimal value 128: 1=water present, 0=no water if all other bits zero
bit 6 - cloud Decimal value: 64: 1=pixel masked out due to cloud, 0=unmasked
bit 5 - cloud shadow Decimal value: 32: 1=pixel masked out due to cloud shadow,
0=unmasked
bit 4 - high slope Decimal value: 16: 1=pixel masked out due to high slop, 0=unmasked
bit 3 - terrain shadow Decimal value: 8: 1=pixel masked out due to terrain shadow,
0=unmasked
bit 2 - sea
Decimal value: 4: 1=pixel masked out due to being over sea,
0=unmasked
bit 1 - contiguity
Decimal value: 2: 1=pixel masked out due to lack of data contiguity,
0=unmasked
bit 0 - no data
Decimal value: 1:1=pixel masked out due to NO_DATA in NBAR
source, 0=valid data in NBAR
'''


def unpack_wofs(data):
    """Unpack an array of WOFS data to a collection of binary arrays
    of the same shape."""
    return namedtuple('WOFS', [
            'water_present', 'cloud', 'cloud_shadow', 'high_slope',
            'terrain_shadow', 'sea', 'contiguity', 'nodata'])(
        *np.unpackbits(data.astype(np.uint8)[np.newaxis], axis=0).astype(bool))    
```
